# Importing necessary libraries and dataset, establishing database connection

The analytical process begins with the importation of essential libraries: sqlite3 for database interactions and pandas for data manipulation and analysis. The nycflights13 dataset, encompassing various aspects of flight information, is loaded into Pandas DataFrames. Subsequent to the data loading phase, a connection to a SQLite database is established. This connection is pivotal for performing SQL-based queries and operations on the dataset. The integration of the sqlite3 library enables the creation of a relational database environment, where each CSV file from the nycflights13 dataset is transformed into a distinct database table. Moreover, In the below code comment='#' is used to skip the comment part present in the dataset, So basically, these comments will not be read into the DataFrame.


```python
#Importing the necessary libraries 
import sqlite3
import pandas as pd

# Loading the datasets
AK_flightsNYCdf = pd.read_csv('../Dataset/nycflights13_flights_csv.gz',compression='gzip', delimiter=',', header=0, comment='#')
AK_airlinesNYCdf = pd.read_csv('../Dataset/nycflights13_airlines_csv.gz',compression='gzip', delimiter=',', header=0, comment='#')
AK_airportsNYCdf = pd.read_csv('../Dataset/nycflights13_airports_csv.gz',compression='gzip', delimiter=',', header=0, comment='#')
AK_planesNYCdf = pd.read_csv('../Dataset/nycflights13_planes_csv.gz',compression='gzip', delimiter=',', header=0, comment='#')
AK_weatherNYCdf = pd.read_csv('../Dataset/nycflights13_weather_csv.gz',compression='gzip', delimiter=',', header=0, comment='#')

# Establishing a connection to a new SQLite database
conn = sqlite3.connect('nycflights13byAK.db')

# Exporting CSV files to the SQLite database
AK_flightsNYCdf.to_sql('flights', conn, if_exists='replace', index=False)
AK_airlinesNYCdf.to_sql('airlines', conn, if_exists='replace', index=False)
AK_airportsNYCdf.to_sql('airports', conn, if_exists='replace', index=False)
AK_planesNYCdf.to_sql('planes', conn, if_exists='replace', index=False)
AK_weatherNYCdf.to_sql('weather', conn, if_exists='replace', index=False)
```

### Query 1: Distinct Engines in Planes

In the below code, the first part of the code involves executing a SQL query to extract unique values from the engine column in the planes table. This is achieved using the pd.read_sql_query function from the pandas library, which allows for the execution of SQL queries directly in Python.The SQL query "SELECT DISTINCT engine FROM planes" is passed as a parameter to this function. The SELECT DISTINCT statement is used to retrieve only the unique values from the engine column. The result of this query is stored in the variable query1_sql_byAK, which becomes a pandas DataFrame containing the unique engine types found in the planes table of the connected SQL database (referenced by conn).

The second part of the code provides a pandas equivalent of the SQL query. Here, the objective is to extract the unique values from the engine column of the DataFrame AK_planesNYCdf,this is achieved using the drop_duplicates() method on the engine column of AK_planesNYCdf. The method drop_duplicates() returns a Series with duplicate values removed, thus yielding a list of unique engine types. The result is stored in query1_pd_byAK.

The final part of the code is a validation step to ensure that the outputs of the SQL query and the pandas operation are identical. The function pd.testing.assert_frame_equal is used to assert that the two DataFrames query1_sql_byAK and query1_pd_byAK are equal. Several transformations are applied to both DataFrames to standardize their structure for that pupose sort_values('engine') and reset_index('drop=True') for both DataFrames, ensuring that they are not only identical in terms of data but also aligned in their ordering. This process guarantees that the comparison is accurate, considering both the contents and the sequence of the data.  


```python
# Written in sql query version
query1_sql_byAK = pd.read_sql_query("SELECT DISTINCT engine FROM planes", conn)

# Written in pandas Version 
query1_pd_byAK = AK_planesNYCdf['engine'].drop_duplicates()

# Testing both the query if they are matching or not using pd testing assert frame
pd.testing.assert_frame_equal(query1_sql_byAK.sort_values('engine').reset_index(drop=True), query1_pd_byAK.to_frame().sort_values('engine').reset_index(drop=True))
```

### Query 3: Count of Planes Grouped by Engine

The initial part of the code executes a SQL query to count the number of occurrences of each distinct engine type in the planes table. This is facilitated through the pd.read_sql_query function, which is a component of the pandas library and allows for the integration of SQL queries within a Python environment. The specific SQL query "SELECT COUNT(*), engine FROM planes GROUP BY engine" is passed as a parameter to this function. The GROUP BY clause groups the records by the engine column, and COUNT(*) calculates the number of records in each group. This results in a tally of each engine type's occurrences in the dataset. The outcome of this query is stored in query3_sql_byAK, a pandas DataFrame, which encapsulates the count of each engine type as found in the connected SQL database (referred to by conn).

Now talking about pandas version which provides an equivalent operation using pandas. The goal is to replicate the SQL query's functionality by counting the occurrences of each engine type within the DataFrame AK_planesNYCdf. This DataFrame presumably contains data similar to the planes table of the SQL database. The operation is performed using the groupby('engine').size() method chain on AK_planesNYCdf. The groupby method groups the data by the engine column, and size() computes the size of each group. The result is a pandas Series, which is then converted into a DataFrame with reset_index(name='COUNT(*)'), providing a column with the count of each engine type. The DataFrame query3_pd_byAK is then rearranged to match the column order of the SQL query result, ensuring that the first column represents the counts and the second column represents the engine types.

The concluding part of the code is focused on verifying the equivalence of the results obtained from the SQL query and the pandas computation for that pupose we again use the pd.testing.assert_frame_equal function which employed to assert the equality of query3_sql_byAK and query3_pd_byAK DataFrames. This function performs a comparison of both the data and the structure of the two DataFrames.


```python
# Written in sql query version
query3_sql_byAK = pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine", conn)

# Written in pandas Version 
query3_pd_byAK = AK_planesNYCdf.groupby('engine').size().reset_index(name='COUNT(*)')

# Reorder columns to match the SQL result for pandas version
query3_pd_byAK = query3_pd_byAK[['COUNT(*)', 'engine']]

# Testing both the query if they are matching or not using pd testing assert frame
pd.testing.assert_frame_equal(query3_sql_byAK, query3_pd_byAK)
```

### Query 5: Aggregates on Plane Year by Engine and Manufacturer

The first part of the code involves executing a SQL query to aggregate and analyze the year data for each combination of engine and manufacturer in the planes table. This is accomplished with the pd.read_sql_query function from the pandas library. The query "SELECT MIN(year), AVG(year), MAX(year), engine, manufacturer FROM planes GROUP BY engine, manufacturer" is used. This SQL command computes the minimum (MIN), average (AVG), and maximum (MAX) values of the year column, grouping the results by both engine and manufacturer. The resulting DataFrame, query5_sql_byAK, contains these aggregated values for each engine and manufacturer combination as found in the SQL database, which is referred to by conn.

The subsequent segment of the code offers a pandas-based approach to achieve the same aggregation. The aim here is to compute the minimum, average, and maximum values of the year column for each engine and manufacturer grouping within the DataFrame AK_planesNYCdf. The aggregation is performed using the groupby(['engine', 'manufacturer']).agg() method. The agg() function specifies the aggregation operations to be performed on the year column: minimum, mean, and maximum. The result is a multi-level column DataFrame, which is then flattened and renamed to align with the SQL query's output format. Column names are set to 'MIN(year)', 'AVG(year)', and 'MAX(year)', matching the SQL result column names.

After renaming the columns, the DataFrame query5_pd_byAK undergoes further transformation to match the column order of the SQL query result. The columns are rearranged so that the aggregation results (minimum, average, and maximum year) come first, followed by engine and manufacturer. 

The final step in the code is to test the equivalence of the DataFrames obtained from the SQL and pandas methods. The pd.testing.assert_frame_equal function is used for this purpose. This function compares query5_sql_byAK and query5_pd_byAK to check if they are identical in terms of both data and structure. 


```python
# Written in sql query version
query5_sql_byAK = pd.read_sql_query("SELECT MIN(year), AVG(year), MAX(year), engine, manufacturer FROM planes GROUP BY engine, manufacturer", conn)

# Written in pandas Version 
query5_pd_byAK = AK_planesNYCdf.groupby(['engine', 'manufacturer']).agg({
    'year': ['min', 'mean', 'max']
}).reset_index()

# Rearrange and rename columns in the Pandas DataFrame to match sql dataframe names
query5_pd_byAK.columns = [("year_min" if col[1] == 'min' else "year_mean" if col[1] == 'mean' else "year_max" if col[1] == 'max' else col[0]) for col in query5_pd_byAK.columns]
query5_pd_byAK.columns = [
    'engine',
    'manufacturer',
    'MIN(year)',
    'AVG(year)',
    'MAX(year)'
]

# Rearrange the columns of the Pandas DataFrame to match the SQL query results
query5_pd_byAK = query5_pd_byAK[
    ['MIN(year)', 'AVG(year)', 'MAX(year)', 'engine', 'manufacturer']
]

# Test for equality of the entire DataFrames
pd.testing.assert_frame_equal(query5_sql_byAK, query5_pd_byAK)
```

### Query 12: Manufacturers with More Than 10 Planes

In the initial segment of the code, a SQL query is executed to count the number of planes for each manufacturer in the planes table, but only including those manufacturers with more than 10 planes. This is accomplished using the pd.read_sql_query function from the pandas library, enabling SQL query execution in a Python environment. The SQL statement "SELECT manufacturer, COUNT(*) FROM planes GROUP BY manufacturer HAVING COUNT(*) > 10" is employed. The GROUP BY clause groups the records by manufacturer, and the COUNT(*) function calculates the total count of planes for each group. The HAVING clause is a key element here, as it filters out the groups (manufacturers) that do not meet the condition of having more than 10 planes. The resulting DataFrame, query12_sql_byAK, thus includes only those manufacturers with a count exceeding 10, along with the respective counts.

Following the SQL part, the code provides a pandas equivalent for the same task. The objective is to count the number of planes for each manufacturer in the DataFrame AK_planesNYCdf (which mirrors the planes table in the SQL database), but only for those manufacturers with more than 10 planes. This is achieved using a two-step process in pandas. First, the groupby('manufacturer').filter(lambda x: len(x) > 10) method chain is used. The groupby method groups the data by manufacturer, and the filter method with a lambda function filters out those groups with 10 or fewer planes. The resulting DataFrame is then regrouped by manufacturer, and size() computes the count of planes for each manufacturer. The reset_index(name='COUNT(*)') transformation turns the result into a DataFrame with a column named 'COUNT(*)' representing the count for each manufacturer. The DataFrame query12_pd_byAK thus represents the count of planes for each manufacturer, excluding those with 10 or fewer planes.

Again pd.testing.assert_frame_equal function is used to assert the equality of the two DataFrames, query12_sql_byAK and query12_pd_byAK. Prior to the comparison, both DataFrames are sorted by the manufacturer column, and their indices are reset (dropping the old index) using sort_values('manufacturer').reset_index(drop=True). These steps ensure that the DataFrames are compared not only based on their data content but also on their manufacturers' order.


```python
# Written in sql query version
query12_sql_byAK = pd.read_sql_query("SELECT manufacturer, COUNT(*) FROM planes GROUP BY manufacturer HAVING COUNT(*) > 10", conn)

# Written in pandas Version
query12_pd_byAK = AK_planesNYCdf.groupby('manufacturer').filter(lambda x: len(x) > 10).groupby('manufacturer').size().reset_index(name='COUNT(*)')

# Testing both the query if they are matching or not using pd testing assert frame
pd.testing.assert_frame_equal(query12_sql_byAK.sort_values('manufacturer').reset_index(drop=True), query12_pd_byAK.sort_values('manufacturer').reset_index(drop=True))
```

### Query 15: Joining Flights and Planes Data

The SQL version of query 15 (query15_sql_byAK) demonstrates a relational database operation where data from two different tables, flights and planes, are integrated. This integration is achieved through a LEFT JOIN operation, a fundamental SQL technique used to combine rows from two or more tables based on a related column between them. The query specifically selects all columns from the flights table and certain columns from the planes table (namely year, speed, and seats), with the year, speed, and seats columns being renamed to plane_year, plane_speed, and plane_seats, respectively, for clarity. This renaming aids in distinguishing between similar column names in the joined tables. The join condition is based on the tailnum column, which serves as a common identifier between flights and planes. The resulting DataFrame from this operation provides a comprehensive view of flight details combined with specific attributes of the planes used for those flights.

In the pandas version of the query, a similar merging operation is replicated using DataFrame methods. The merging of the AK_flightsNYCdf (representing the flights table) and a subset of AK_planesNYCdf (representing the planes table) is performed using pd.merge. This method is analogous to the SQL LEFT JOIN and is specified as such with the how='left' parameter. The merge operation in pandas also utilizes the tailnum column as the key for joining. To ensure consistency with the SQL query's output, the columns from the planes DataFrame are selectively included and subsequently renamed to match the SQL version's aliases. This renaming is crucial for maintaining parity between the two data sources and for providing an unambiguous understanding of each column's context.

Before testing for equivalence, both query15_sql_byAK and query15_pd_byAK are sorted by the tailnum column and their indices are reset (dropping the old indices). This standardizes the row order in both DataFrames, making them comparable. The pd.testing.assert_frame_equal function is then used to assert the equality of these sorted DataFrames, query15_sql_sorted and query15_pd_sorted.


```python
# SQL Version
query15_sql_byAK = pd.read_sql_query("SELECT flights.*, planes.year AS plane_year, planes.speed AS plane_speed, planes.seats AS plane_seats FROM flights LEFT JOIN planes ON flights.tailnum = planes.tailnum", conn)

# Pandas Version
query15_pd_byAK = pd.merge(AK_flightsNYCdf, AK_planesNYCdf[['tailnum', 'year', 'speed', 'seats']], how='left', left_on='tailnum', right_on='tailnum')

# Renaming the columns of the Pandas DataFrame to match the SQL query result
query15_pd_byAK = query15_pd_byAK.rename(columns={'year_x': 'year', 'year_y': 'plane_year', 'speed': 'plane_speed', 'seats': 'plane_seats'})

# Now, sort and reset index for both DataFrames before testing for equality
query15_sql_sorted = query15_sql_byAK.sort_values(by='tailnum').reset_index(drop=True)
query15_pd_sorted = query15_pd_byAK.sort_values(by='tailnum').reset_index(drop=True)

# Testing for equality of the entire DataFrames
pd.testing.assert_frame_equal(query15_sql_sorted, query15_pd_sorted)

```

### Query 16: Joining Flights, Planes, and Airlines Data

In the SQL portion of the code, a complex query involving a subquery and multiple inner joins is executed. The pd.read_sql_query function in pandas is used to run the SQL command "SELECT planes.*, airlines.* FROM (SELECT DISTINCT carrier, tailnum FROM flights) AS cartail INNER JOIN planes ON cartail.tailnum = planes.tailnum INNER JOIN airlines ON cartail.carrier = airlines.carrier". This query first creates a subquery (SELECT DISTINCT carrier, tailnum FROM flights) to select unique combinations of carrier and tailnum from the flights table. The resulting temporary table cartail is then joined with the planes table on tailnum and subsequently with the airlines table on carrier. The final output, query16_sql_byAK, is a DataFrame that combines columns from both planes and airlines tables for each unique pair of carrier and tailnum from the flights table.

In pandas, the same process is replicated in a step-by-step manner. First, a subquery equivalent is created by selecting distinct carrier and tailnum combinations from AK_flightsNYCdf using the drop_duplicates() method. This DataFrame distinct_carrier_tailnum then undergoes two successive inner joins. The first join merges distinct_carrier_tailnum with AK_planesNYCdf on tailnum, and the second join merges the result of the first join with AK_airlinesNYCdf on carrier. The resulting DataFrame query16_pd_byAK combines information from planes and airlines based on the unique pairs of carrier and tailnum from flights.

Before testing for equivalence, query16_pd_byAK is reordered to ensure its column order matches that of the SQL query result. The DataFrames are then sorted by tailnum and carrier to align the row order. The pd.testing.assert_frame_equal function is used to assert the equality of query16_sql_sorted and query16_pd_sorted after resetting their indices (dropping the old indices). This comparison checks if the DataFrames are identical in terms of both data content and structure.


```python
# SQL Version
query16_sql_byAK = pd.read_sql_query("SELECT planes.*, airlines.* FROM (SELECT DISTINCT carrier, tailnum FROM flights) AS cartail INNER JOIN planes ON cartail.tailnum = planes.tailnum INNER JOIN airlines ON cartail.carrier = airlines.carrier", conn)

# Create a subquery equivalent: Distinct 'carrier' and 'tailnum' from flights
distinct_carrier_tailnum = AK_flightsNYCdf[['carrier', 'tailnum']].drop_duplicates()

# First join: Merge the 'distinct_carrier_tailnum' with 'planes' on 'tailnum'
join_with_planes = pd.merge(distinct_carrier_tailnum, AK_planesNYCdf, on='tailnum', how='inner')

# Second join: Merge the result of the first join with 'airlines' on 'carrier'
query16_pd_byAK = pd.merge(join_with_planes, AK_airlinesNYCdf, on='carrier', how='inner')

# Ensure the column order is the same as in SQL query result
# Reorder columns: Start with 'tailnum' and then columns from 'planes' followed by 'airlines'
query16_pd_byAK = query16_pd_byAK[['tailnum', 'year', 'type', 'manufacturer', 'model', 'engines', 'seats',
                                   'speed', 'engine', 'carrier', 'name']]
# Sorting both DataFrames by 'tailnum' and 'carrier' for a consistent comparison
query16_sql_sorted = query16_sql_byAK.sort_values(by=['tailnum', 'carrier']).reset_index(drop=True)
query16_pd_sorted = query16_pd_byAK.sort_values(by=['tailnum', 'carrier']).reset_index(drop=True)

# Testing for equality of the entire DataFrames
pd.testing.assert_frame_equal(query16_sql_sorted, query16_pd_sorted)
```

### Query 17: Joining Filtered Flights Data with Aggregated Weather Data

The execution of query 17 involves a sophisticated SQL operation and its corresponding implementation in pandas, designed to integrate flight and weather data. In the SQL version (query17_sql_byAK), conducted through the pd.read_sql_query function in pandas, a complex query is crafted. This query, articulated in SQL, begins with a subquery that selects all records from the flights table where the flights originate from 'EWR' (Newark Liberty International Airport). A second subquery is simultaneously formulated to compute the daily average temperature and humidity from the weather table, exclusively for 'EWR'. These subqueries are then adeptly joined using a LEFT JOIN on the year, month, and day columns, thereby merging flight data with corresponding daily weather statistics. The final outcome is a comprehensive DataFrame that aligns each flight from 'EWR' with the day's average temperature and humidity.

Parallel to the SQL query, the pandas version (query17_pd_byAK) undertakes a similar analytical approach but through a stepwise approach native to pandas operations. It starts by filtering the AK_flightsNYCdf DataFrame to include only those flights departing from 'EWR'. Next, a strategic aggregation is performed on the weather data (from AK_weatherNYCdf) for 'EWR', focusing on calculating mean values for temperature and humidity, day-wise. These mean values are aptly renamed to atemp and ahumid for clarity and consistency with the SQL query's output. The final and crucial step involves merging the filtered flight data with the aggregated weather statistics. This merge, executed as a left join, aligns the flight data with the relevant daily weather metrics, echoing the logic applied in the SQL query.

To ensure the validity of these parallel methodologies, both the SQL and pandas DataFrames are first sorted by year, month, day, and tailnum to standardize the row order, followed by a reset of their indices. This ordering and resetting are pivotal for an equitable comparison. Subsequently, the pd.testing.assert_frame_equal function is invoked to assert the equivalence of these sorted DataFrames. 

Concluding the operation with the closure of the SQL database connection (conn) signifies the end of data retrieval and analysis.

```python
# SQL Version
query17_sql_byAK = pd.read_sql_query("""
SELECT flights2.*, atemp, ahumid FROM 
    (SELECT * FROM flights WHERE origin='EWR') AS flights2 
LEFT JOIN 
    (SELECT year, month, day, AVG(temp) AS atemp, AVG(humid) AS ahumid FROM weather WHERE origin='EWR' GROUP BY year, month, day) AS weather2 
ON flights2.year=weather2.year AND flights2.month=weather2.month AND flights2.day=weather2.day
""", conn)

# Pandas Version
# Filter flights data
ewr_flights_pd_byAK = AK_flightsNYCdf[AK_flightsNYCdf['origin'] == 'EWR']

# Aggregate weather data
ewr_weather_agg_pd_byAK = AK_weatherNYCdf[AK_weatherNYCdf['origin'] == 'EWR'].groupby(['year', 'month', 'day']).agg({'temp': 'mean', 'humid': 'mean'}).rename(columns={'temp': 'atemp', 'humid': 'ahumid'}).reset_index()

# Merge flights with weather data
query17_pd_byAK = pd.merge(ewr_flights_pd_byAK, ewr_weather_agg_pd_byAK, how='left', on=['year', 'month', 'day'])

query17_sql_sorted = query17_sql_byAK.sort_values(by=['year', 'month', 'day', 'tailnum']).reset_index(drop=True)
query17_pd_sorted = query17_pd_byAK.sort_values(by=['year', 'month', 'day', 'tailnum']).reset_index(drop=True)

# Testing for equality of the entire DataFrames
pd.testing.assert_frame_equal(query17_sql_sorted, query17_pd_sorted)
```
